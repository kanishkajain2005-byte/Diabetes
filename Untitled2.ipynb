Lets handle the zero values of insulin, BloodPressure ,SkinThickness, Glucose and BMI

```python
insulin_median = data[data['Insulin'] != 0]['Insulin'].median()
data['Insulin'].replace(0, insulin_median, inplace=True)
BloodPressure_median = data[data['BloodPressure'] != 0]['BloodPressure'].median()
data['BloodPressure'].replace(0, BloodPressure_median, inplace=True)
SkinThickness_median = data[data['SkinThickness'] != 0]['SkinThickness'].median()
data['SkinThickness'].replace(0, SkinThickness_median, inplace=True)
BMI_median = data[data['BMI'] != 0]['BMI'].median()
data['BMI'].replace(0, BMI_median, inplace=True)
Glucose_median = data[data['Glucose'] != 0]['Glucose'].median()
data['Glucose'].replace(0, Glucose_median, inplace=True)
```

```python
print(data['Insulin'])
print(data['BloodPressure'])
print(data['BMI'])
print(data['SkinThickness'])
print(data['Glucose'])
```

# Data visualisation

### Finding the relation of Age groups and the Chances of having Diabetes.

```python
bins = [20,25,30,35,40,45,50,55,60,65,70,75,80,85]
labels = ['20-25','25-30','30-35','35-40','40-45','45-50','50-55','55-60','60-65','65-70','70-75','75-80','80-85']
data['Age'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='Outcome', data=data)
plt.title('Proportion of Diabetes Cases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Proportion of Diabetes')
plt.show()
```

As we can see from the bar graph that the chances of having Diabetes increases with the age and it is most common for the people of age 30-60, 70-75 and the heighest chase is for 50-55 age group.

```python
plt.figure(figsize = (12,12))
for i, col in enumerate(['Pregnancies','Glucose','BloodPressure','BMI','DiabetesPedigreeFunction','Outcome']):
    plt.subplot(3,3,i+1)
    sns.histplot(x= col, data=data, kde= True)
plt.show()
```

The above graph tells us about the most common values of the dataset features.

Ex -
*   For Pregnancies its 0 and 2.
*   For Outcome its only 1 and 0.






      

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Insulin'], bins=10, kde=True, edgecolor='black')
plt.title('Proportion of Diabetes Cases by Insulin')
plt.xlabel('Insulin')
plt.ylabel('Proportion of Diabetes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['SkinThickness'], bins=10, kde=True, edgecolor='black')
plt.title('Proportion of Diabetes Cases by SkinThickness')
plt.xlabel('SkinThickness')
plt.ylabel('Proportion of Diabetes')
plt.show()
```

Here is the relation bw skinthickness and chances of having diabetes.
